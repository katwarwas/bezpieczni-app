```python
import folium
import pandas as pd
from geopy.geocoders import Nominatim
import time
from collections import Counter

geolocator = Nominatim(user_agent="bezpieczni-app", timeout=10)

map = folium.Map(location=[0, 0], zoom_start=2, tiles='CartoDB dark_matter')

# Unique countries in the 'country' column
unique_countries = filtered_df['country'].unique()

# Add points to the map based on data from the data frame
for country in unique_countries:
    try:
        country_code = country_codes.get(country, 'brak kodu')
        location = geolocator.geocode(country, country_codes=country_code)
        country_polish = country_names_pl.get(country, 'Tłumaczenie niedostępne')
        if location:
            country_data =filtered_df[filtered_df['country'] == country]
            event_type_counts = Counter(country_data['event_type'])
            popup_text = "<br>".join([f"{event_type}: {count}" for event_type, count in event_type_counts.items()])
            folium.Marker(
                [location.latitude, location.longitude], 
                popup=f"{country_polish} - typ ataku:<br>{popup_text}"
            ).add_to(map)
    except Exception as e:
        print(f"Could not geocode {country}: {e}")
    print(country_code)
    print(country)

# Save the map to an HTML file
# map.save('../static/map2/map-type.html')
map
```
