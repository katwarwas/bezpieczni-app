```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Tworzenie figury i osi
fig, ax = plt.subplots(figsize=(21, 15))

# Dodawanie aktorw jako ikony
ax.text(0.1, 0.9, 'Go\n(Zwyky uytkownik)', horizontalalignment='center', verticalalignment='center', fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))
ax.text(0.1, 0.5, 'Moderator', horizontalalignment='center', verticalalignment='center', fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))
ax.text(0.1, 0.1, 'Administrator', horizontalalignment='center', verticalalignment='center', fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))

# Dodawanie przypadkw uycia jako elipsy
cases = [
    ('Przegldanie informacji o cyberatakach', 0.5, 0.9),
    ('Przegldanie aktualnoci', 0.5, 0.8),
    ('Przegldanie analiz', 0.5, 0.7),
    ('Przegldanie mapy atakw', 0.5, 0.6),
    ('Przegldanie linkw', 0.5, 0.5),
    ('Dodawanie artykuu', 0.5, 0.4),
    ('Edycja wasnego artykuu', 0.5, 0.3),
    ('Usuwanie wasnego artykuu', 0.5, 0.2),
    ('Zarzdzanie uytkownikami', 0.5, 0.1),
    ('Edycja dowolnego artykuu', 0.5, 0.05),
    ('Usuwanie dowolnego artykuu', 0.5, 0)
]

for case in cases:
    ax.text(case[1], case[2], case[0], horizontalalignment='center', verticalalignment='center', fontsize=12, bbox=dict(facecolor='lightblue', edgecolor='black', boxstyle='round,pad=1'))

# Linie czce aktorw z przypadkami uycia
lines = [
    [(0.1, 0.9), (0.5, 0.9)],
    [(0.1, 0.9), (0.5, 0.8)],
    [(0.1, 0.9), (0.5, 0.7)],
    [(0.1, 0.9), (0.5, 0.6)],
    [(0.1, 0.9), (0.5, 0.5)],
    [(0.1, 0.5), (0.5, 0.4)],
    [(0.1, 0.5), (0.5, 0.3)],
    [(0.1, 0.5), (0.5, 0.2)],
    [(0.1, 0.1), (0.5, 0.1)],
    [(0.1, 0.1), (0.5, 0.4)],
    [(0.1, 0.1), (0.5, 0.05)],
    [(0.1, 0.1), (0.5, 0)]
]

for line in lines:
    ax.plot([line[0][0], line[1][0]], [line[0][1], line[1][1]], 'k-', linewidth=1)

# Ustawienia osi i tytuu
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
ax.set_title('Przypadki uycia dla aplikacji "Bezpieczni"', fontsize=16)

plt.savefig('przypadki_uycia')

# Wywietlanie diagramu
plt.show()
```
