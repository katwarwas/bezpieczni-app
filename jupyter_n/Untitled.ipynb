```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Tworzenie figury i osi
fig, ax = plt.subplots(figsize=(21, 15))

# Dodawanie aktorów jako ikony
ax.text(0.1, 0.9, 'Gość\n(Zwykły użytkownik)', horizontalalignment='center', verticalalignment='center', fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))
ax.text(0.1, 0.5, 'Moderator', horizontalalignment='center', verticalalignment='center', fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))
ax.text(0.1, 0.1, 'Administrator', horizontalalignment='center', verticalalignment='center', fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))

# Dodawanie przypadków użycia jako elipsy
cases = [
    ('Przeglądanie informacji o cyberatakach', 0.5, 0.9),
    ('Przeglądanie aktualności', 0.5, 0.8),
    ('Przeglądanie analiz', 0.5, 0.7),
    ('Przeglądanie mapy ataków', 0.5, 0.6),
    ('Przeglądanie linków', 0.5, 0.5),
    ('Dodawanie artykułu', 0.5, 0.4),
    ('Edycja własnego artykułu', 0.5, 0.3),
    ('Usuwanie własnego artykułu', 0.5, 0.2),
    ('Zarządzanie użytkownikami', 0.5, 0.1),
    ('Edycja dowolnego artykułu', 0.5, 0.05),
    ('Usuwanie dowolnego artykułu', 0.5, 0)
]

for case in cases:
    ax.text(case[1], case[2], case[0], horizontalalignment='center', verticalalignment='center', fontsize=12, bbox=dict(facecolor='lightblue', edgecolor='black', boxstyle='round,pad=1'))

# Linie łączące aktorów z przypadkami użycia
lines = [
    [(0.1, 0.9), (0.5, 0.9)],
    [(0.1, 0.9), (0.5, 0.8)],
    [(0.1, 0.9), (0.5, 0.7)],
    [(0.1, 0.9), (0.5, 0.6)],
    [(0.1, 0.9), (0.5, 0.5)],
    [(0.1, 0.5), (0.5, 0.4)],
    [(0.1, 0.5), (0.5, 0.3)],
    [(0.1, 0.5), (0.5, 0.2)],
    [(0.1, 0.1), (0.5, 0.1)],
    [(0.1, 0.1), (0.5, 0.4)],
    [(0.1, 0.1), (0.5, 0.05)],
    [(0.1, 0.1), (0.5, 0)]
]

for line in lines:
    ax.plot([line[0][0], line[1][0]], [line[0][1], line[1][1]], 'k-', linewidth=1)

# Ustawienia osi i tytułu
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
ax.set_title('Przypadki użycia dla aplikacji "Bezpieczni"', fontsize=16)

plt.savefig('przypadki_użycia')

# Wyświetlanie diagramu
plt.show()
```
