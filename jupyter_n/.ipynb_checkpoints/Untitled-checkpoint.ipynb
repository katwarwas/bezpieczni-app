```python
import pandas as pd
import numpy as np 
df = pd.read_csv('cyberattacks_database.csv')
#checking data
df.head()
```

```python
#checking latest data
df.tail()
```

```python
#cleaning data 
df = df[df['event_date'] != "1969-12-31"]
df.tail()
```

```python
unique_countries = df['country'].unique()
# sorted_countries = sorted(unique_countries)
# sorted_countries
# check 
for country in unique_countries:
    if not isinstance(country, str):
        print('country:', country )
    

filtered_df = df[~df['country'].apply(lambda x: not isinstance(x, str))]
filtered_df

unique_countries = filtered_df['country'].unique()
sorted_countries = sorted(unique_countries)
sorted_countries
```

```python
most_attacked_country = {}
for country in sorted_countries:
    count = (df['country'] == country).sum()
    if count > 100:
        most_attacked_country[country] = count
print(most_attacked_country)
most_attacked_country.pop('Undetermined', None)
most_attacked_country
```

```python
translations = {
    'Australia': 'Australia',
    'Brazil': 'Brazylia',
    'Canada': 'Kanada',
    'China': 'Chiny',
    'France': 'Francja',
    'Germany': 'Niemcy',
    'India': 'Indie',
    'Israel': 'Izrael',
    'Italy': 'Włochy',
    'Japan': 'Japonia',
    'Poland': 'Polska',
    'Russian Federation': 'Federacja Rosyjska',
    'Ukraine': 'Ukraina',
    'United Kingdom of Great Britain and Northern Ireland': 'Wielka Brytania',
    'United States of America': 'Stany Zjednoczone Ameryki'
}
most_attacked_country_polish = {translations[country]: count for country, count in most_attacked_country.items() if country in translations}
most_attacked_country_polish_sorted = dict(sorted(most_attacked_country_polish.items()))
most_attacked_country_polish_sorted
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch



fig, ax = plt.subplots(figsize=(15,12))
fig.patch.set_facecolor('#292e47')

plt.subplots_adjust(bottom=0.25)

labels = list(most_attacked_country_polish_sorted.keys())  
values = list(most_attacked_country_polish_sorted.values())
bars = ax.bar(labels, values, color='cornflowerblue')

ax.set_frame_on(False)
ax.set_ylabel('Liczba ataków', color='white', fontsize=20)
ax.set_title('Najczęściej atakowane kraje', color='white', fontsize=20, pad=40)

ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, color='white', rotation=45, ha='right', fontsize=15)
ax.tick_params(axis='x', colors='white', labelsize=16)
ax.tick_params(axis='y', colors='white', labelsize=12)

ax.set_ylim(0, 1000)

for bar in bars:
    yval = bar.get_height()
    if yval < 1000:
        ax.text(bar.get_x() + bar.get_width()/2, yval, f'{yval}', va='bottom', ha='center', color='white')
    else: 
        ax.text(bar.get_x() + bar.get_width()/2, 960, f'{yval}', va='bottom', ha='center', color='white')

plt.savefig('most_attaked_country')
plt.show()
```

```python
from datetime import datetime
from datetime import datetime
date_str= np.array(filtered_df['event_date'])

# Funkcja do przekształcenia stringa w datę
def string_to_date(date_str):
    return datetime.strptime(date_str, "%Y-%m-%d").date()

# Wektorowe przekształcenie całej tablicy
vectorized_string_to_date = np.vectorize(string_to_date)
dates = vectorized_string_to_date(date_str)
dates
```

```python
year_event = dates.astype('datetime64[Y]')
year_events = year_event.astype('str')
year_events_name = np.unique(year_events)
year_events_count = pd.DataFrame(pd.DataFrame(year_events).value_counts()).sort_index().values
```

```python
from matplotlib.ticker import MultipleLocator

fig, ax = plt.subplots(figsize=(15,12))
fig.patch.set_facecolor('#292e47')
ax.plot(year_events_name, year_events_count, marker = 'o', color='cornflowerblue', linewidth=3, markersize=15, linestyle='--')

plt.subplots_adjust(bottom=0.1)

ax.set_frame_on(False)

ax.set_xlabel('Rok', color='white', fontsize=14, labelpad=20)
ax.set_ylabel('Liczba ataków', color='white', fontsize=12, labelpad=20)
ax.tick_params(axis='both', colors='white')
ax.set_ylim(0, 3000)

plt.title('Liczba ataków w danym roku', fontsize=20, color='white', pad=40)

# fig.subplots_adjust(left=0.2, right=0.8, top=0.8, bottom=0.2)
plt.savefig('number_of_events_by_year')
plt.show()
```

```python
import math
motive = dict(filtered_df['motive'].value_counts()[filtered_df['motive'].value_counts() > 1]) 
motive

translations = {
    'Financial': 'Finansowy',
    'Undetermined': 'Nieokreślony',
    'Protest': 'Protest',
    'Political-Espionage': 'Szpiegostwo polityczne',
    'Sabotage': 'Sabotaż',
    'Personal Attack': 'Atak osobisty',
    'Industrial-Espionage': 'Szpiegostwo przemysłowe',
    'Espionage': 'Szpiegostwo'
}


motive_pl = {translations[key]: value for key, value in motive.items()}

total = sum(motive_pl.values())
percentages = [f"{(value / total) * 200/ 2:.1f}%" for value in motive_pl.values()]



labels = [f"{key} - {value}" for key, value in zip(motive_pl.keys(), percentages)]

labels
```

```python
fig, ax = plt.subplots(figsize=(15,15))
fig.patch.set_facecolor('#292e47')

def make_autopct(values):
    def my_autopct(pct):
        value = int(round(pct*total/100.0))
        rounded_pct = math.floor(pct * 2) / 2
        return f'{pct:.1f}%' if value > total * 0.01 else ''  
    return my_autopct

ax.pie(motive.values(), pctdistance=0.8, autopct=make_autopct(motive.values()), colors=['#3b466d', '#455388', '#4f63a6', '#5a74b6',  '#6d8bc4', '#89a6d1', '#afc5e1', '#d0dced'], startangle=45, textprops={'color':'white', 'fontsize':14})

circle = plt.Circle((0, 0), 0.60, fc='#292e47')
ax.add_artist(circle)
plt.title('Motywy Cyberataków', fontsize=20, color='white' )
plt.legend(labels, loc='center', framealpha=0.3, fontsize=16, labelcolor = 'white')
plt.savefig('cyber_attack_motives')
plt.show()
```
